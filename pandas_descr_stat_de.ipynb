### Daten aus der CSV-Datei lesen

Nun benutzen wir pandas um die [CSV-Datei](https://de.wikipedia.org/wiki/CSV_(Dateiformat)) einzulesen. Damit das funktioniert muss vorher noch das pandas-Paket mit dem `import`-Befehl unter dem Namen `pd` verfügbar gemacht werden.

Dabei wird ein sogenannter [DataFrame](https://pandas.pydata.org/docs/reference/api/pandas.DataFrame.html) mit dem Namen `nba` erzeugt. Vereinfacht kann man sich darunter eine Tabelle mit Spalten und Zeilen vorstellen. DataFrames sind **die** zentralen Datentypen in pandas.   

Mit dem Befehl `type()` kann der Datentyp eines jeden Python-Objekts ausgegeben werden. 

```python
import pandas as pd
import matplotlib.pyplot as plt
nba = pd.read_csv("nba_all_elo.csv")
type(nba)
pd.set_option("display.max.columns", None)
# Expected:
# <class 'pandas.core.frame.DataFrame'>
```

```python
import seaborn as sns

sns.set_style("whitegrid")
```

```python
plt.figure(figsize=(10,10))
sns.scatterplot(data=nba[(nba["_iscopy"]==0)& ((nba["team_id"]=="LAL") & (nba["opp_id"]== "GSW"))], x="pts", y="opp_pts", hue="game_result", size="difference")
```

```python

sns.displot(
    nba[(nba["_iscopy"]==0)&(nba["year_id"]>2010) & ((nba["team_id"]=="GSW") | (nba["team_id"]== "BOS"))], x="pts", col="fran_id", row="game_result",
    binwidth=5, height=4, facet_kws=dict(margin_titles=True),
)
```

```python
plt.figure(figsize=(30,10))
ax = sns.boxplot(data=nba[(nba["_iscopy"]==0)&(nba["year_id"]==2015)], x="fran_id", y="pts")

#Die Zeile unterhalb können Sie aktivieren in dem Sie das #-Zeichen löschen. Nun werden die einzelnen Werte als Punkte auf die Boxplots gelegt.
#ax = sns.swarmplot(data=nba[(nba["_iscopy"]==0)&(nba["year_id"]==2015)], x="fran_id", y="pts", color = "0.25")
```
