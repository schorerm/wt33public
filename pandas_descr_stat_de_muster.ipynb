Nun benutzen wir pandas um die [CSV-Datei](https://de.wikipedia.org/wiki/CSV_(Dateiformat)) einzulesen. Damit das funktioniert muss vorher noch das pandas-Paket mit dem `import`-Befehl unter dem Namen `pd` verfügbar gemacht werden.

## Lösung:

### Teilaufgabe 1

In wie vielen Spielen erzielte eine der Mannschaften weniger als 50 Punkte? In wie vielen mehr als 150 Punkte?

```python
nba[(nba["pts"]>150) | (nba["pts"]<50)]
```

```python
nba[(nba["pts"]>150)].game_id.count()
```

```python
nba[(nba["pts"]<50)].game_id.count()
```

Achtung! Spiele können bei der obigen Abfrage zweimal gezählt werden (beide Mannschaften erzielen < 50 bzw. > 150), da Spiele bekanntlich aus Sicht beider Mannschaften aufgeführt sind. 

Eine Alternative wäre (mit Abfrage des Feldes `_iscopy`):

```python
#Spieleliste:
nba[ (nba["_iscopy"] == 0) & 
( (nba["pts"]>150) | (nba["pts"]<50) |(nba["opp_pts"]>150) | (nba["opp_pts"]<50) ) ]
```

```python
#Anzahl:
nba[(nba["_iscopy"] == 0) & ((nba["pts"]>150)| (nba["opp_pts"]>150))].game_id.count()
```

```python
#Anzahl
nba[(nba["_iscopy"] == 0) & ((nba["pts"]<50)| (nba["opp_pts"]<50))].game_id.count()
```

### Teilaufgabe 4

```python
nba[(nba["team_id"]=="GSW") &
(nba["year_id"]>=2000)
].pts.mean()
```

### Teilaufgabe 6

### Teilaufgabe 7

Erzeugen Sie die neue Spalte `difference` im DataFrame nba. Diese soll die Differenz zwischen der Spalte `pts`und der Spalte `opp_pts` enthalten.
Visualisieren Sie die Häufigkeitsverteilung der Punktedifferenzen der Spiele (ohne Kopien -> `is_copy == 0`) mit einem Histogram.


```python
nba["difference"] = nba.pts - nba.opp_pts
nba[nba["_iscopy"]==0].difference.plot(kind="hist")
```

### Teilaufgabe 8

- 8.1 Der Scatterplot zeigt alle Heimspiele der Los Angeles Lakers gegen die Golden State Warriors. An der y-Achse sind die Punkte der Lakers, an der y-Achse die Punkte der Warriors abgebildet. Sieg oder Niederlage der Lakers ist durch die Farbe der Punkte ausgedrückt. Die Höhe des Punktunterschieds ist durch die Größe der Punkte dargestellt. 
- 8.2 Die displot-Funktion (https://seaborn.pydata.org/generated/seaborn.displot.html) generiert für das angegebene DataSet (alle Spiele der Boston Celtics und der Golden State Warriors ohne Kopien) eine Reihe von Sub-Visualisierungen. Hierbei werden Histogramme für erzielten Punkte für jede Mannschaft nch Spielresultat getrennt dargestellt. Es werden hierbei so viele Visualisierungen generiert wie Kombinationen von Ausprägungen in den Daten vorhanden sind. Hier sind es 2 (Mannschaften) x 2 (mögl. Spielresultate). 
- 8.3 Die boxplot-Funktion von seaborn kann die Datenreihe automatisch nach Mannschaft (`fran_id`) aufteilen und die Boxplots für die erzielten Punkte für 2015 übersichtlich nebeneinander darstellen.  
- 8.4 Hier sieht man eine weitere mögliche Anwendung der oben beschriebenen Funktion. Es wird die in Aufgabe 7 ermittelte Punktedifferenz in allen Partien eines Jahres in einem Boxplot pro Jahr für den gesamten Zeitraum dargestellt.
